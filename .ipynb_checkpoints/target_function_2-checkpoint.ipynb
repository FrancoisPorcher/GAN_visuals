```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow.keras.layers as layers
from tensorflow.keras.models import Model
from tensorflow.keras.optimizers import Adam
import os
```

# First function

```python
def target_function(Z):
    '''
    Map the noise Z in [0,1] to a point (z, sin(z)) in [0, 2pi]x[-1, 1]
    '''
    theta = Z[:, 0]*  np.pi / 2
    x1 = np.cos(theta)
    x2 = np.sin(theta)
    results = np.zeros((Z.shape[0], 2))
    results[:, 0] = x1
    results[:, 1] = x2
    return results
```

```python
Z = np.linspace(0, 1, 1000).reshape(-1, 1)
real_sample = target_function(Z)
fig = plt.figure(figsize = (5, 5))
plt.title("Distribution we are trying to learn")
plt.xlabel("x1")
plt.ylabel("x2")
plt.scatter(real_sample[:, 0], real_sample[:, 1], c='r', label='real')
```

# Defining the gans

```python
NUM_BATCHES = 6000
BATCH_SIZE = 1024
PLOT_EVERY = 100
GRID_RESOLUTION = 400
z_dim = 1
```

```python
def generate_noise(samples):
    '''
    Generate `samples` samples of uniform noise in 
    ([0,1])
    '''
    return np.random.uniform(0, 1, (samples, z_dim))
```

```python
def sample_from_target_function(samples):
    '''
    This generates a random sample from the target_distriution we are trying to learn
    '''
    Z = generate_noise(samples)
    return target_function(Z)
```

```python
def build_generator():
    '''
    Build a generator mapping (R, R) to ([-1,1], [-1,1])
    '''
    input_layer = layers.Input((z_dim,))
    X = input_layer
    for i in range(5):
        X = layers.Dense(128)(X)
        X = layers.LeakyReLU(0.1)(X)
        X = layers.Dropout(0.1)(X)
    output_layer = layers.Dense(2)(X)
    G = Model(input_layer, output_layer)
    return G


def build_discriminator():
    '''
    Build a discriminator mapping (R, R) to [0, 1]
    '''
    input_layer = layers.Input((2,))
    X = input_layer
    for i in range(3):
        X = layers.Dense(128)(X)
        X = layers.LeakyReLU(0.1)(X)
        X = layers.Dropout(0.1)(X)
    output_layer = layers.Dense(1, activation='sigmoid')(X)
    D = Model(input_layer, output_layer)
    D.compile(Adam(learning_rate=0.001, beta_1=0.5),
                        loss='binary_crossentropy',
                        metrics=['accuracy'])
    return D

def build_GAN(G, D):
    '''
    Given a generator and a discriminator, build a GAN
    '''
    D.trainable = False
    input_layer = layers.Input((z_dim,))
    X = G(input_layer)
    output_layer = D(X)
    GAN = Model(input_layer, output_layer)
    GAN.compile(Adam(learning_rate=0.0002, beta_1=0.5),
                        loss='binary_crossentropy',
                        metrics=['accuracy'])
    return GAN
```

```python
grid = np.zeros((GRID_RESOLUTION, GRID_RESOLUTION, 2))
grid[:, :, 0] = np.linspace(0, 1, GRID_RESOLUTION).reshape((1, -1))
grid[:, :, 1] = np.linspace(1, 0, GRID_RESOLUTION).reshape((-1, 1))
flat_grid = grid.reshape((-1, 2))
```

```python
test_samples = sample_from_target_function(1000)
test_noise = generate_noise(1000)

X = np.linspace(0, 1, 400)
X = X.reshape(-1, 1)
real_samples = target_function(X)
```

```python
def plot_losses(G, D, GAN, step, step_count, D_accuracy, D_loss, G_accuracy, G_loss, filename):
    '''
    Plots for the GAN gif
    '''
    f, ax = plt.subplots(figsize=(5,5))
    plt.title("Losses, Epoch: {}".format(step))

    # [0, 0]: plot loss and accuracy
    ax.plot(step_count, 
                    G_loss,
                    label='G loss',
                    c='darkred',
                    zorder=50,
                    alpha=0.8,)
    
    ax.plot(step_count, 
                    G_accuracy,
                    label='G accuracy',
                    c='lightcoral',
                    zorder=40,
                    alpha=0.8,)
    ax.plot(step_count, 
                    D_loss,
                    label='D loss',
                    c='darkblue',
                    zorder=55,
                    alpha=0.8,)
    ax.plot(step_count, 
                    D_accuracy,
                    label='D accuracy',
                    c='cornflowerblue',
                    zorder=45,
                    alpha=0.8,)
    
    ax.set_xlim(-5, NUM_BATCHES+5)
    ax.set_ylim(-0.1, 2.1)
    ax.legend(loc=1)


    plt.tight_layout()
    plt.show()
    plt.savefig(filename)
    plt.close()
    
    
def plot_distribution(G, D, GAN, step, step_count, D_accuracy, D_loss, G_accuracy, G_loss, filename):
    '''
    Plots for the GAN gif
    '''

    fake_samples = G.predict(generate_noise(200), batch_size=200)
    fig = plt.figure(figsize=(7, 7))
    plt.scatter(fake_samples[:, 0], fake_samples[:, 1], edgecolor='red', facecolor='None', s=5, alpha=0.7, linewidth=1, label='fake')
    plt.scatter(real_samples[:, 0],real_samples[:, 1], edgecolor='blue', facecolor='None', s=5, alpha=0.7, linewidth=1, label = 'real')
    plt.title("Generated samples, epoch {}".format(step))
    plt.legend()

    plt.tight_layout()
    plt.show()
    plt.savefig(filename)
    plt.close()
    
def plot_region(G, D, GAN, step, step_count, D_accuracy, D_loss, G_accuracy, G_loss, filename):
    '''
    Plots for the GAN gif
    '''

    fig, ax = plt.subplots(figsize=(10, 10))
    plt.title("Region Decision of the discriminator, Epoch {}".format(step), fontsize=20)
    ax.set_xticks(np.arange(0, GRID_RESOLUTION+1, GRID_RESOLUTION//4))
    ax.set_xticklabels(np.linspace(0, 1, 5))
    ax.set_yticks(np.arange(0, GRID_RESOLUTION+1, GRID_RESOLUTION//4))
    ax.set_yticklabels(np.linspace(0, 1, 5))
    plt.xlabel("x1", fontsize=20)
    plt.ylabel("x2", fontsize=20)
    confidences = D.predict(flat_grid, batch_size=BATCH_SIZE)
    confidences = confidences.reshape((GRID_RESOLUTION, GRID_RESOLUTION))

    ax.imshow(confidences, vmin=0, vmax=1, cmap='seismic_r')


    plt.tight_layout()
    plt.savefig(filename)
    plt.close()

## Set up directories
paths = ['ims', 
            'ims/target_function_1', 
                'ims/target_function_1/samples','ims/target_function_1/losses', 'ims/target_function_1/confidence_region', 'ims/target_function_1/models',
            'ims/target_function_2', 
                'ims/target_function_2/samples','ims/target_function_2/losses','ims/target_function_2/confidence_region','ims/target_function_2/models',
            'ims/target_function_3', 
                'ims/target_function_3/samples','ims/target_function_3/losses','ims/target_function_3/confidence_region','ims/target_function_3/models',
        ]
for i in paths:
    if not os.path.exists(i):
        os.makedirs(i)
```

```python
## Part 3: GAN
G = build_generator()
D = build_discriminator()
GAN = build_GAN(G, D)

step_count = []
D_accuracy = []
G_accuracy = []
D_loss = []
G_loss = []
count = 0
for step in range(NUM_BATCHES):
    # Train discriminator
    D.trainable = True
    real_data = sample_from_target_function(BATCH_SIZE // 2)
    fake_data = G.predict(generate_noise(BATCH_SIZE // 2), batch_size=BATCH_SIZE // 2)
    data = np.concatenate((real_data, fake_data), axis=0)
    real_labels = np.ones((BATCH_SIZE // 2, 1))
    fake_labels = np.zeros((BATCH_SIZE // 2, 1))
    labels = np.concatenate((real_labels, fake_labels), axis=0)
    _D_loss, _D_accuracy = D.train_on_batch(data, labels)


    # Train generator
    D.trainable = False
    noise = generate_noise(BATCH_SIZE)
    labels = np.ones((BATCH_SIZE, 1))
    _G_loss, _G_accuracy = GAN.train_on_batch(noise, labels)

    
    if step % PLOT_EVERY == 0:
        print("Current step: {}".format(step))
        step_count.append(step)
        D_loss.append(_D_loss)
        D_accuracy.append(_D_accuracy)
        G_loss.append(_G_loss)
        G_accuracy.append(_G_accuracy)
        
        plot_losses(G=G, 
             D=D,
             GAN=GAN,
             step=step,
             step_count=step_count,
             D_accuracy=D_accuracy,
             D_loss=D_loss,
             G_accuracy=G_accuracy,
             G_loss=G_loss,
             filename=f'ims/target_function_2/losses/loss.{count:03d}.png')
        
        plot_distribution(G=G, 
             D=D,
             GAN=GAN,
             step=step,
             step_count=step_count,
             D_accuracy=D_accuracy,
             D_loss=D_loss,
             G_accuracy=G_accuracy,
             G_loss=G_loss,
             filename=f'ims/target_function_2/samples/samples.{count:03d}.png')
        
        plot_region(G=G, 
             D=D,
             GAN=GAN,
             step=step,
             step_count=step_count,
             D_accuracy=D_accuracy,
             D_loss=D_loss,
             G_accuracy=G_accuracy,
             G_loss=G_loss,
             filename=f'ims/target_function_2/confidence_region/confidence_region.{count:03d}.png')
        
        
        count += 1
        
        
        

os.system(f'ffmpeg -r 20 -i ims/3/3.%03d.png'
              f' -crf 15 ims/3/3.mp4')
G.save
G.save("ims/target_function_2/models/G.h5")
D.save("ims/target_function_2/models/D.h5")
```
