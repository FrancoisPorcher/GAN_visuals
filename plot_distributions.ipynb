```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow.keras.layers as layers
from tensorflow.keras.models import Model
from tensorflow.keras.optimizers import Adam
import os
```

# Target Function 1

```python
def target_function(Z):
    '''
    Map the noise Z in [0,1] to a point (z, sin(z)) in [0, 2pi]x[-1, 1]
    '''
    x1 = Z[:, 0] * 2 * np.pi
    x2 = np.sin(x1)
    results = np.zeros((Z.shape[0], 2))
    results[:, 0] = x1
    results[:, 1] = x2
    return results
```

```python
Z = np.linspace(0, 1, 1000).reshape(-1, 1)

real_sample = target_function(Z)
fig = plt.figure(figsize = (10, 5))
plt.title("Distribution we are trying to learn: (x1, sin(x1))")
plt.xlabel("x1")
plt.ylabel("x2")
plt.scatter(real_sample[:, 0], real_sample[:, 1], c='r', label='real')
```

# Target Function 2

```python
def target_function(Z):
    '''
    Map the noise Z in [0,1] to a point (z, sin(z)) in [0, 2pi]x[-1, 1]
    '''
    theta = Z[:, 0]*  np.pi / 2
    x1 = np.cos(theta)
    x2 = np.sin(theta)
    results = np.zeros((Z.shape[0], 2))
    results[:, 0] = x1
    results[:, 1] = x2
    return results
```

```python
Z = np.linspace(0, 1, 1000).reshape(-1, 1)
real_sample = target_function(Z)
fig = plt.figure(figsize = (10, 10))
plt.title("Distribution 2 we are trying to learn: (cos(theta), sin(theta))")
plt.xlabel("x1")
plt.ylabel("x2")

plt.scatter(real_sample[:, 0], real_sample[:, 1], c='r', label='real')
```

# Target Function 3

```python
def target_function(Z):
    '''
    Map the noise Z in [0,1] to a point (z, sin(z)) in [0, 2pi]x[-1, 1]
    '''
    theta = Z[:, 0] *  2 * np.pi
    x1 = theta
    x2 = np.cos(2 * theta)
    results = np.zeros((Z.shape[0], 2))
    results[:, 0] = x1
    results[:, 1] = x2
    return results
```

```python
Z = np.linspace(0, 1, 1000).reshape(-1, 1)
real_sample = target_function(Z)
fig = plt.figure(figsize = (10, 10))
plt.title("Distribution 3 we are trying to learn: (theta, cos(2 theta))")
plt.xlabel("x1")
plt.ylabel("x2")

plt.scatter(real_sample[:, 0], real_sample[:, 1], c='r', label='real')
```
